```python
missed_description = red_river[red_river['Description'].str.contains(r"\bmiss\w*\b|not\s+emptied|not\s+picked\s+up|no\s+pick\s+up", case=False, na=False)]
missed_description
```

```python
missed_categories = ['Trash - Backdoor', 'Trash - Curbside/Alley Missed Pickup']
```

```python
missed_by_category = red_river[red_river['Request'].isin(missed_categories)]
```

```python
missed_red_river = pd.concat([missed_by_category, missed_by_description]).drop_duplicates()
missed_red_river
```
